```
def producer():
    ''' Data received per unit time by zerodha api.  
    This a simulation of original function.
    Every unit of time (example per second), new data(row) will be added to streaming_stocks_data.csv
    '''
    
    with open('stocks_data_2020_till_2020_09_03.csv', 'r') as input_file, open('streaming_stocks_data.csv', 'a', newline='') as output_file:        
        input_file_reader = csv.reader(input_file)
        results = filter(lambda column: column[-1] == '14' and column[1]>='2020-08-01' and column[1]<='2020-08-31', input_file_reader)
        writer = csv.writer(output_file)
        #writer.writerow(['timestamp', 'last_price' , 'volume' , 'scrip_id'])
        for row in results:
            outdata = [row[1], row[5], row[6], row[7]]
            writer.writerow(outdata)
            output_file.flush()
            print(outdata)
            time.sleep(interval_time)

    input_file.close()
    output_file.close()
```

```
producer()
```
