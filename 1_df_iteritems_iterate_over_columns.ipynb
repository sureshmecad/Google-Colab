### **df.items**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'species': ['mammal', 'mammal', 'fish'],
                   'population': [3948, 4000, 6000]},
                  index=['tiger', 'fox', 'shark'])
df
```

1.JPG

```python
for label, content in df.items():
    print('label:', label)
    print('content:', content, sep='\n')
```
