## Bar (Horizontal & Vertical)

Bar graphs are visualizations that show numerical values by category.
By default, the bar uses a rectangular shape.

There are basically two types of bar graphs. Horizontal and vertical version.

The horizontal version is used when there are too many categories.

```python
fig = plt.figure(figsize=(15, 7))

axes = [0, 0]
axes[0] = fig.add_subplot(1,2,1,aspect=1)
axes[1] = fig.add_subplot(1,2,2, aspect=1)

labels = list('ABCDE')
x = np.arange(len(labels)) 
y1 = np.array([1, 2, 3, 4, 3])
y2 = np.array([4, 3, 2, 5, 1])
width=0.5

axes[0].bar(x, y1, 
       color='#1ABDE9',
       edgecolor='black',
       linewidth=2,
       width=width,)


axes[1].barh(x, y1, 
       color='#1ABDE9',
       edgecolor='black',
       linewidth=2,
       height=width,
)


for idx, ax in enumerate(axes, 0):
    
    if idx==0:
        ax.set_xticks(x)
        ax.set_xticklabels(labels, fontsize=12)
        ax.tick_params(axis='y', which='major', labelsize=13)
    else:
        ax.set_yticks(x)
        ax.set_yticklabels(labels, fontsize=12)
        ax.tick_params(axis='x', which='major', labelsize=13)
        
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)


axes[0].set_title(f"Bar Plot (vertical)", fontsize=20, fontweight='semibold')    
axes[1].set_title(f"Bar Plot (horizontal)", fontsize=20, fontweight='semibold')    
    

plt.subplots_adjust(wspace=0.2)
plt.show()
```

## Bar Plot (Space setting)

You can improve the readability of the bar graph by adjusting the spacing and space of the bar.

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 7))

labels = list('ABCDE')
x = np.arange(len(labels)) 
y1 = np.array([1, 2, 3, 4, 3])
y2 = np.array([4, 3, 2, 5, 1])
width=0.5

# Default
axes[0].bar(x, y1, 
       color='#1ABDE9',
)

# Original
axes[1].bar(x, y1, 
       color='#1ABDE9',
       edgecolor='black',
       linewidth=2,
       width=width,)


# Spine
for idx, ax in enumerate(axes, 1):
    ax.set_title(f"Bar Plot {idx}", fontsize=20, fontweight='semibold')
    ax.set_xticks(x)
    ax.set_xticklabels(labels, fontsize=12)
    ax.tick_params(axis='y', which='major', labelsize=13)

axes[0].spines['top'].set_visible(True)
axes[0].spines['right'].set_visible(True)    
axes[1].margins(0.05,0.2)    
axes[1].spines['bottom'].set_linewidth(2)
axes[1].spines['left'].set_linewidth(2)    

plt.subplots_adjust(wspace=0.2)
plt.show()
```

## Multiple Bar Plot

One way to show various features is to draw two or more plots in succession.

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 7), sharey=True)

labels = list('ABCDE')
x = np.arange(len(labels)) 
y1 = np.array([1, 2, 3, 4, 3])
y2 = np.array([4, 3, 2, 5, 1])
width=0.5

# Sky Bar Plot
axes[0].bar(x, y1, 
       color='#1ABDE9',
       edgecolor='black',
       linewidth=2,
       width=width,
       label='sky'
      )

# Pink Bar Plot
axes[1].bar(x, y2, 
       color='#F36E8E',
       edgecolor='black',
       linewidth=2,
       width=width,
       label='pink'
      )


for idx, ax in enumerate(axes, 1):

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_linewidth(2)
    ax.spines['left'].set_linewidth(2)

    # Title
    ax.set_title(f"Bar Plot {idx}", fontsize=20, fontweight='semibold')

    # Label
    ax.set_xlabel("X axis label", fontsize=15, fontweight='medium')
    if idx==1 : ax.set_ylabel("Y axis label", fontsize=15, fontweight='medium')

    # Ticklabels
    ax.set_xticks(x)
    ax.set_xticklabels(labels, fontsize=12)
    ax.tick_params(axis='y', which='major', labelsize=13)

    # Legend 
    ax.legend(fontsize=13)

plt.tight_layout()
plt.show()
```

## Stacked, Overlapped, Grouped Barplot

There are three ways to show more than one feature on a plot:

- Stacked
- Overlapped
- Grouped

### Stacked

As a way to express it in a stack, you need to pay attention to the following factors.

- Are all the bars in the same order?
- Is it appropriate to see the distribution of the bars other than the bottom?
- What are the advantages over overlapping

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

labels = list('ABCDE')
x = np.arange(len(labels)) 
y1 = np.array([1, 2, 3, 4, 3])
y2 = np.array([4, 3, 2, 5, 1])
width=0.5

# Sky Bar Plot
ax.bar(x, y1, 
       color='#1ABDE9',
       edgecolor='black',
       linewidth=2,
       width=width,
       label='sky'
      )

# Pink Bar Plot
ax.bar(x, y2, bottom=y1, 
       color='#F36E8E',
       edgecolor='black',
       linewidth=2,
       width=width,
       label='pink'
      )


# Spine
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

#### Text #### 

# Title
ax.set_title("Stacked Bar Plot", fontsize=20, fontweight='semibold')

# Label
ax.set_xlabel("X axis label", fontsize=15, fontweight='medium')
ax.set_ylabel("Y axis label", fontsize=15, fontweight='medium')

# Ticklabels
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12)
ax.tick_params(axis='y', which='major', labelsize=13)

# Legend 
ax.legend(fontsize=13)

plt.tight_layout()
plt.show()
```

### Percentage Staked Bar Chart

To see the total percentage, you can also draw a bar graph with the total as 1 as follows.

```python
fig = plt.figure(figsize=(10, 5), dpi=150)
ax = fig.add_subplot(111)

labels = list('ABCDE')
x = np.arange(len(labels)) 
y1 = np.array([1, 2, 3, 4, 3])
y2 = np.array([4, 3, 2, 5, 1])
width=0.5

# Sky Bar Plot
ax.barh(x, y1/(y1+y2), 
       color='#1ABDE9',
       edgecolor='black',
       linewidth=2,
       label='sky'
      )

# Pink Bar Plot
ax.barh(x, y2/(y1+y2), left=y1/(y1+y2), 
       color='#F36E8E',
       edgecolor='black',
       linewidth=2,
       label='pink'
      )


# Spine
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

#### Text #### 

# Title
ax.set_title("Percentage Stacked Bar Chart", fontsize=15, fontweight='semibold')

for xx, yy1, yy2 in zip(x, y1, y2):
    ax.annotate(f"{100*yy1/(yy1+yy2):.4}%", 
                   xy=(yy1/(yy1+yy2)/2, xx),
                   va = 'center', ha='center',fontsize=10, fontweight='bold',
                   color='white')

    ax.annotate(f"{100*yy2/(yy1+yy2):.4}%", 
               xy=((1+(yy1)/(yy1+yy2))/2, xx),
               va = 'center', ha='center',fontsize=10, fontweight='bold',
               color='white')
    
# Ticklabels
ax.set_xticks([])
ax.set_yticks(x)
ax.set_yticklabels(labels, fontsize=12)
ax.tick_params(axis='y', which='major', labelsize=13)

# Legend 
ax.legend(loc='lower center', ncol=3, bbox_to_anchor=(0.5, -0.06))

plt.tight_layout()
plt.show()
```

### Overlapped 

It is suitable for comparison, which was a disadvantage in a stacked bar chart.

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

labels = list('ABCDE')
x = np.arange(len(labels)) 
y1 = np.array([1, 2, 3, 4, 3])
y2 = np.array([4, 3, 2, 5, 1])
width=0.5

# Sky Bar Plot
ax.bar(x, y1, 
       color='#1ABDE9',
       edgecolor='black',
       linewidth=2,
       width=width,
       label='sky',
       alpha=0.8
      )

# Pink Bar Plot
ax.bar(x, y2,
       color='#F36E8E',
       edgecolor='black',
       linewidth=2,
       width=width,
       label='pink',
       alpha=0.8
      )


# Spine
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

#### Text #### 

# Title
ax.set_title("Overlapped Bar Plot", fontsize=20, fontweight='semibold')

# Label
ax.set_xlabel("X axis label", fontsize=15, fontweight='medium')
ax.set_ylabel("Y axis label", fontsize=15, fontweight='medium')

# Ticklabels
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12)
ax.tick_params(axis='y', which='major', labelsize=13)

# Legend 
ax.legend(fontsize=13)

plt.tight_layout()
plt.show()
```

### Grouped

This is the best way to look at a variety of features. It is characterized by being adjacent to each other and is best when the number of colors is 5-7.

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

labels = list('ABCDE')
x = np.arange(len(labels)) 
y1 = np.array([1, 2, 3, 4, 3])
y2 = np.array([4, 3, 2, 5, 1])
width=0.3

# Sky Bar Plot
ax.bar(x-width/2, y1, 
       color='#1ABDE9',
       edgecolor='black',
       linewidth=2,
       width=width,
       label='sky'
      
      )

# Pink Bar Plot
ax.bar(x+width/2, y2, 
       color='#F36E8E',
       edgecolor='black',
       linewidth=2,
       width=width,
       label='pink'
      )


# Spine
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

#### Text #### 

# Title
ax.set_title("Grouped Bar Plot", fontsize=20, fontweight='semibold')

# Label
ax.set_xlabel("X axis label", fontsize=15, fontweight='medium')
ax.set_ylabel("Y axis label", fontsize=15, fontweight='medium')

# Ticklabels
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12)
ax.tick_params(axis='y', which='major', labelsize=13)

# Legend 
ax.legend(fontsize=13)

plt.tight_layout()
plt.show()
```

## Line Plot

Line plots are visualized with points on the data and the lines connecting those points.

```python
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

fig = plt.figure(figsize=(7, 5), dpi=150)




np.random.seed(970725)

x = np.arange(20)
y = np.random.rand(20)

# Ax2
ax2 = fig.add_subplot(111)
ax2.plot(x, y,
        marker='o',
       color='#1ABDE9',
       linewidth=2,)

ax2.spines['bottom'].set_linewidth(1.5)
ax2.spines['left'].set_linewidth(1.5)
ax2.set_title(f"Line Plot", loc='left', fontsize=12, va= 'bottom', fontweight='semibold')

plt.tight_layout()
plt.show()
```

## Simple is Better (Line Ver.)


If you want to see a lot of information, it is better to draw it complex, and if you want to know trends rather than information, it is better to draw simple.

Always consider the difference between EDA and Presentation.

```python
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

fig = plt.figure(figsize=(12, 5), dpi=150)




np.random.seed(970725)

x = np.arange(20)
y = np.random.rand(20)


# Ax1
ax1 = fig.add_subplot(121)
ax1.plot(x, y,
       color='#1ABDE9',
         marker='o',
       linewidth=2, label='sin')

# Ax2
ax2 = fig.add_subplot(122)
ax2.plot(x, y,
       color='#1ABDE9',
       linewidth=2,)

ax2.spines['bottom'].set_linewidth(1.5)
ax2.spines['left'].set_linewidth(1.5)

ax1.yaxis.set_major_locator(MultipleLocator(0.05))    
ax1.xaxis.set_major_locator(MultipleLocator(1))    
ax1.grid(linewidth=0.3)    
ax1.set_title(f"Line Plot (information)", loc='left', fontsize=12, va= 'bottom', fontweight='semibold')
ax2.set_title(f"Line Plot (clean)", loc='left', fontsize=12, va= 'bottom', fontweight='semibold')

plt.tight_layout()
plt.show()
```

## Line Plot (Label & Legend)

It is better to use labels instead of legends.

```python
fig = plt.figure(figsize=(10, 8), dpi=150)


x = np.linspace(0, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# Ax1
ax1 = fig.add_subplot(211, aspect=1)
ax1.plot(x, y1,
       color='#1ABDE9',
       linewidth=2, label='sin')

ax1.plot(x, y2,
       color='#F36E8E',
       linewidth=2, label='cos')

# Ax2
ax2 = fig.add_subplot(212, aspect=1)
ax2.plot(x, y1,
       color='#1ABDE9',
       linewidth=2,)

ax2.plot(x, y2,
       color='#F36E8E',
       linewidth=2,)

ax2.text(x[-1]+0.1, y1[-1], s='sin', fontweight='bold',
         va='center', ha='left', 
         bbox=dict(boxstyle='round,pad=0.3', fc='#1ABDE9', ec='black', alpha=0.3))

ax2.text(x[-1]+0.1, y2[-1], s='cos', fontweight='bold',
         va='center', ha='left', 
         bbox=dict(boxstyle='round,pad=0.3', fc='#F36E8E', ec='black', alpha=0.3))



for ax in [ax1, ax2]:
    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

    
    
ax1.legend(loc='upper center')
ax1.set_title(f"Line Plot (Legend)", loc='left', fontsize=12, va= 'bottom', fontweight='semibold')
ax2.set_title(f"Line Plot (Label w/o Legend)", loc='left', fontsize=12, va= 'bottom', fontweight='semibold')

plt.tight_layout()
plt.show()
```

## Line Plot (add Max, Minor)

Add information through various elements for necessary content.

```python
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)

fig = plt.figure(figsize=(12, 5), dpi=150)




np.random.seed(970725)

x = np.arange(20)
y = np.random.rand(20)


# Ax1
ax1 = fig.add_subplot(121)
ax1.plot(x, y,
       color='#1ABDE9',
       linewidth=2, label='sin')


# Ax2
ax2 = fig.add_subplot(122)
ax2.plot(x, y,
       color='lightgray',
       linewidth=2,)

ax2.scatter(x[np.argmax(y)], np.max(y), 
            c='#F36E8E',s=50, zorder=20)

ax2.scatter(x[np.argmin(y)], np.min(y), 
            c='#1ABDE9',s=50, zorder=20)

for ax in [ax1, ax2]:
    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

ax2.set_ylim(-0.1, 1.1)
ax2.set_xlim(-1, 20)

ax2.text(x[np.argmax(y)], np.max(y)+0.04, s=f'{np.max(y):.3}', fontweight='bold',
         va='bottom', ha='center', 
         bbox=dict(boxstyle='round,pad=0.3', fc='#F36E8E', ec='black', alpha=0.3))

ax2.text(x[np.argmin(y)], np.min(y)-0.04, s=f'{np.min(y):.3}', fontweight='bold',
         va='top', ha='center', 
         bbox=dict(boxstyle='round,pad=0.3', fc='#1ABDE9', ec='black', alpha=0.3))


ax2.margins(0)
plt.hlines(np.max(y), xmin=-1, xmax=x[np.argmax(y)], linestyle='--', color='#F36E8E')
plt.hlines(np.min(y), xmin=-1, xmax=x[np.argmin(y)], linestyle='--', color='#1ABDE9')

ax1.set_title(f"Default", loc='left', fontsize=12, va= 'bottom', fontweight='semibold')
ax2.set_title(f"Add Min/Max Info", loc='left', fontsize=12, va= 'bottom', fontweight='semibold')


plt.tight_layout()
plt.show()
```

## Scatter Plot

The scatterplot can display X-axis/Y-axis data simultaneously in a Cartesian coordinate system. In other words, it basically visualizes two-dimensional data.

Since one point represents one data, it is possible to express the data quantitatively and the distribution.

```python
fig = plt.figure(figsize=(7, 7), dpi=150)
ax = fig.add_subplot(111, aspect=1)

np.random.seed(970725)

x = np.random.rand(20)
y = np.random.rand(20)

ax.scatter(x, y, s=150, 
           c='#1ABDE9',
           linewidth=1.5,
           edgecolor='black')

ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)

ax.set_xlim(0, 1.1)
ax.set_ylim(0, 1.1)

ax.set_title(f"Basic Scatter Plot", fontsize=15,va= 'center', fontweight='semibold')


plt.tight_layout()
plt.show()
```

## Scatter with/without Grid

When the points are small, the grid may not be useful, so be careful.

```python
fig = plt.figure(figsize=(16, 7), dpi=150)
np.random.seed(970725)

x = np.random.rand(20)
y = np.random.rand(20)

axes = []
for i in range(1, 3):
    ax = fig.add_subplot(1, 2, i, aspect=1)
    axes.append(ax)
    ax.scatter(x, y, s=150, 
               c='#1ABDE9',
               linewidth=1.5,
               edgecolor='black', zorder=10)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

    ax.set_xlim(0, 1.1)
    ax.set_ylim(0, 1.1)

    
axes[0].grid(zorder=0)    
axes[0].set_title(f"Scatter Plot w Grid", fontsize=15,va= 'center', fontweight='semibold')
axes[1].set_title(f"Scatter Plot w/o Grid", fontsize=15,va= 'center', fontweight='semibold')

plt.tight_layout()
plt.show()
```

## Scatter Components

The elements of a point have color, marker, and size, and color is best for discrete comparisons.

```python
fig = plt.figure(figsize=(7, 7), dpi=150)
ax = fig.add_subplot(111, aspect=1)

ax.scatter([1, 2, 3], [3, 3, 3], 
           color=['#1ABDE9', '#F36E8E', '#00BE6C'],
           s=300,
           linewidth=1.5,
           edgecolor='black'
          )

for idx, marker in enumerate(['o', '^', 's'], 1):
    ax.scatter(idx, 2,
           color='lightgray',
           marker=marker,
           s=300,
           linewidth=1.5,
           edgecolor='black'
          )

    
for idx, s in enumerate([300, 1200, 2700], 1):
    ax.scatter(idx, 1,
           color='lightgray',
           s=s,
           linewidth=1.5,
           edgecolor='black'
          )    
    
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

ax.set_xlim(0, 4)
ax.set_ylim(0, 3.5)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title(f"Scatter Components", fontsize=15,va= 'center', fontweight='semibold')


plt.tight_layout()
plt.show()
```

## Scatter Components (2)

Using these elements to separate content is called a visual popout. As you can see, you can see the difference in the order of color> shape> size.

```python
fig = plt.figure(figsize=(23, 7), dpi=150)

np.random.seed(970725)
x1 = np.random.rand(10)
y1 = np.random.rand(10)

x2 = np.random.rand(10)
y2 = np.random.rand(10)


# Ax1
ax1 = fig.add_subplot(131, aspect=1)
ax1.scatter(x1, y1, s=150, 
           c='#1ABDE9',
           linewidth=1.5,
           edgecolor='black',
           label='A'
          )

ax1.scatter(x2, y2, s=150, 
           c='#F36E8E',
           linewidth=1.5,
           edgecolor='black',
           label='B')

# Ax2
ax2 = fig.add_subplot(132, aspect=1)
ax2.scatter(x1, y1, s=150, 
           marker='o',
            c='black',
           linewidth=1.5,
           edgecolor='black',
           label='A'
          )

ax2.scatter(x2, y2, s=150, 
           marker='^',
            c='black',
           linewidth=1.5,
           edgecolor='black',
           label='B')

# Ax3
ax3 = fig.add_subplot(133, aspect=1)
ax3.scatter(x1[:3], y1[:3], s=100, 
           marker='o',
           c='#1ABDE9',
           linewidth=1.5,
           edgecolor='black',
           label='100'
          )

ax3.scatter(x1[3:6], y[3:6], s=200, 
           marker='o',
           c='#1ABDE9',
           linewidth=1.5,
           edgecolor='black',
           label='200'
          )

ax3.scatter(x[6:], y[6:], s=300, 
           marker='o',
           c='#1ABDE9',
           linewidth=1.5,
           edgecolor='black',
           label='300'
          )


# Setting
for ax in [ax1, ax2, ax3]:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

    ax.set_xlim(0, 1.1)
    ax.set_ylim(0, 1.1)
    ax.legend(fontsize=10)

ax1.set_title(f"Scatter Plot [Color]", fontsize=15,va= 'center', fontweight='semibold')    
ax2.set_title(f"Scatter Plot [Marker]", fontsize=15,va= 'center', fontweight='semibold')    
ax3.set_title(f"Scatter Plot [Size]", fontsize=15,va= 'center', fontweight='semibold')    
    
plt.tight_layout()
plt.show()
```

## Correlation in Scatter Plot

The scatterplot is basically used to look for correlations.

Correlation is divided into positive correlation, negative correlation, and no relationship.

```python
fig = plt.figure(figsize=(23, 7), dpi=150)

np.random.seed(19970725)
n = 200

x = np.random.normal(0, 0.2, n)
y = np.random.normal(0, 0.2, n)

# Ax1
ax1 = fig.add_subplot(131, aspect=1)
ax1.scatter(x+np.arange(n)/n, y+np.arange(n)/n, s=100, 
           c='#1ABDE9',
           linewidth=1.1,
           edgecolor='black',
           label='A'
          )

# Ax2
ax2 = fig.add_subplot(132, aspect=1)
ax2.scatter(x+np.arange(n)/n, y+(n-np.arange(n)-1)/n, s=100, 
           c='#1ABDE9',
           linewidth=1.1,
           edgecolor='black',
           label='A'
          )

# Ax3
ax3 = fig.add_subplot(133, aspect=1)
ax3.scatter(x*1.5+0.6, y*1.5+0.6, s=100, 
           c='#1ABDE9',
           linewidth=1.1,
           edgecolor='black',
           label='A'
          )

# Setting
for ax in [ax1, ax2, ax3]:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

    ax.set_xlim(-0.5, 1.5)
    ax.set_ylim(-0.5, 1.5)

ax1.set_title(f"Positive Correlation", fontsize=15,va= 'center', fontweight='semibold')    
ax2.set_title(f"Negative Correlation", fontsize=15,va= 'center', fontweight='semibold')    
ax3.set_title(f"No Correlation", fontsize=15,va= 'center', fontweight='semibold')    
    
plt.tight_layout()
plt.show()
```

## Scater Plot's 3 view point 

In addition, you can look at a total of three things.

- **Clusters** : Can the data be divided into similar categories?
- **Gap in Values** : Is there a big gap between the data?
- **Outlier** : Are there any values in the data that are considered odd?

```python
fig = plt.figure(figsize=(23, 7), dpi=150)

np.random.seed(19970725)
n = 200


# Ax1
x = np.random.normal(0, 0.1, n)
y = np.random.normal(0, 0.2, n)

ax1 = fig.add_subplot(131, aspect=1)
ax1.scatter(0.2+x, 1+y, s=100, 
           c='#1ABDE9',
           linewidth=1.1,
           edgecolor='black',
           label='A'
          )

x = np.random.normal(0, 0.15, n)
y = np.random.normal(0, 0.15, n)

ax1.scatter(1+x, 0+y, s=100, 
           color='#F36E8E',
           linewidth=1.1,
           edgecolor='black',
           label='B'
          )

x = np.random.normal(0, 0.25, n)
y = np.random.normal(0, 0.25, n)

ax1.scatter(x, y, s=100, 
           color='#00BE6C',
           linewidth=1.1,
           edgecolor='black',
           label='C'
          )

# Ax2
ax2 = fig.add_subplot(132)

x = np.random.normal(0, 0.06, n)
y = np.random.normal(0, 0.3, n)


ax2.scatter(x, y, s=100, 
           c='#1ABDE9',
           linewidth=1.1,
           edgecolor='black',
          )


x = np.random.normal(0, 0.06, n)
y = np.random.normal(0, 0.3, n)

ax2.scatter(x+1, y, s=100, 
           c='#1ABDE9',
           linewidth=1.1,
           edgecolor='black',
          )

ax2.axvspan(0.25, 0.8, alpha=0.2, color='#1ABDE9')

# Ax3
ax3 = fig.add_subplot(133, aspect=1)

x = np.random.normal(0, 0.2, n)
y = np.random.normal(0, 0.2, n)

ax3.scatter(x, y, s=100, 
           c='lightgray',
           linewidth=1.1,
           edgecolor='black',
          )

ax3.scatter([-0.8, 0.75, 0.8], [-0.7, 0.8, 0.7], s=100,
            c='#F36E8E',
            linewidth=1.1,
            edgecolor='black',
           )

# Setting
for ax in [ax1, ax2, ax3]:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)
    ax.set_xticks([])
    ax.set_yticks([])

ax3.set_xlim(-1, 1)    
ax3.set_ylim(-1, 1)
    
ax1.set_title(f"Cluster", fontsize=15,va= 'center', fontweight='semibold')    
ax2.set_title(f"Gap in Values", fontsize=15,va= 'center', fontweight='semibold')    
ax3.set_title(f"Outliers", fontsize=15,va= 'center', fontweight='semibold')    
    
plt.tight_layout()
plt.show()
```

## Ovecome Overplotting in Scatter

The points on the scatterplot easily overlap. If these points overlap, it is difficult to know the actual distribution, so the following technique can be used.

- Transparency
- Jittering
- 2D Histogram
- Contour plot

```python
from matplotlib.colors import ListedColormap
fig = plt.figure(figsize=(23, 7), dpi=150)

np.random.seed(19970725)
n = 2000

x = np.arange(n)/n+np.random.normal(0, 0.2, n)
y = np.arange(n)/n+np.random.normal(0, 0.2, n)

# Ax1
ax1 = fig.add_subplot(131, aspect=1)
ax1.scatter(x, y, s=50, 
           c='#1ABDE9', 
            alpha=0.3,
            linewidth=0
          )

# Ax2
N = 256
vals = np.ones((N, 4))
vals[:, 0] = np.linspace(1, 26/256, N)
vals[:, 1] = np.linspace(1, 189/256, N)
vals[:, 2] = np.linspace(1, 233/256, N)
newcmp = ListedColormap(vals)

ax2 = fig.add_subplot(132, aspect=1)
ax2.hist2d(x, y, bins=(20, 20), 
           cmap=newcmp)

# Ax3

ax3 = fig.add_subplot(133, aspect=1)
sns.kdeplot(x=x, y=y, ax=ax3, fill=True, cmap=newcmp);

# Setting
for ax in [ax1, ax2, ax3]:
    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['left'].set_linewidth(1.5)

    ax.set_xticks([])
    ax.set_yticks([])
    
    ax.set_xlim(-0.5, 1.5)
    ax.set_ylim(-0.5, 1.5)

ax1.set_title(f"Transparency", fontsize=15,va= 'center', fontweight='semibold')    
ax2.set_title(f"2D Histogram", fontsize=15,va= 'center', fontweight='semibold')    
ax3.set_title(f"Contour Plot", fontsize=15,va= 'center', fontweight='semibold')    
    
plt.tight_layout()
plt.show()
```

## Area can change in Radar Plot

The area of the radar plot does not mean the quality of the actual data.

```python
fig = plt.figure(figsize=(12, 5),dpi=150)


for i in range(2):
    color='#004731'
    ax = fig.add_subplot(1, 2, i+1, projection='polar')



    angles = [np.deg2rad((90+n*60)%360) for n in range(6)]
    angles += angles[:1]

    values=np.array([20, 240, 90, 200, 30, 180])
    labels=np.array(list('ABCDEF'))
    
    if i == 1 :
        labels = labels[np.argsort(values)]
        values.sort()
        
    values = values.tolist()
    values += values[:1]

    

    ax.set_thetagrids([(90+n*60)%360 for n in range(6)], labels, fontweight='bold')
    
    ax.set_rgrids(range(50, 260, 50), range(50, 260, 50), color="gray", size=8)
    ax.set_ylim(0,260)


    ax.plot(angles, values, color= color, linewidth=2, linestyle='solid')
    ax.fill(angles, values, color=color, alpha=0.3)

plt.tight_layout()
plt.show()
```
