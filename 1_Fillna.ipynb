```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                  columns=list("ABCD"))
df
```

##### **1) Replace all NaN elements with 0s**

```python
df.fillna(0)
```

##### **2) Using method Parameter**

- propagate non-null values **forward or backward**

```python
df.fillna(method="ffill")
```

3.JPG

##### **3) Replace all NaN elements in column A, B, C, and D, with 0, 1, 2, and 3 respectively**

```python
values = {"A": 0, "B": 1, "C": 2, "D": 3}

df.fillna(value=values)
```

##### **4) Using Limit**

- Only replace the **first NaN element**

```python
df.fillna(value=values, limit=1)
```

```python
# When filling using a DataFrame, replacement happens along the same column names and same indices

df2 = pd.DataFrame(np.zeros((4, 4)), columns=list("ABCE"))
df.fillna(df2)
```
