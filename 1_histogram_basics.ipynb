### **What is a Histogram?**

- **Histograms** are a way to **visualize** the **distribution of the data.**

- A **histogram** is a chart that uses **bars represent frequencies** which helps visualize **distributions of data.**

- Bars can represent **unique values or groups of numbers** that fall into ranges. The **taller the bar**, the **more data** falls into that range.

- The **shape** of the histogram displays the spread of a **continuous sample of data.**
