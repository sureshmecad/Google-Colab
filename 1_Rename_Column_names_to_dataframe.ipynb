```python
# importing the pandas library
import pandas as pd
```

```python
# creating lists
l1 =["Amar", "Barsha", "Carlos", "Tanmay", "Misbah"]
l2 =["Alpha", "Bravo", "Charlie", "Tango", "Mike"]
l3 =[23, 25, 22, 27, 29]
l4 =[69, 54, 73, 70, 74]
  
# creating the DataFrame
team = pd.DataFrame(list(zip(l1, l2, l3, l4))) 
  
# displaying the DataFrame
print(team)
```

- **Adding column name** to the DataFrame : We can add columns to an existing DataFrame using its **columns** attribute.

```python
# adding column name to the respective columns
team.columns =['Name', 'Code', 'Age', 'Weight']
  
# displaying the DataFrame
print(team)
```

- **Renaming column name** of a DataFrame : We can rename the columns of a DataFrame by using the **rename()** function.

```python
# reanming the DataFrame columns
team.rename(columns = {'Code':'Code-Name', 
                       'Weight':'Weight in kgs'}, 
            inplace = True)
  
# displaying the DataFrame
print(team)
```

```python
df = pd.DataFrame([('Foreign Cinema', 'Restaurant', 289.0),
                   ('Liho Liho', 'Restaurant', 224.0),
                   ('500 Club', 'bar', 80.5),
                   ('The Square', 'bar', 25.30)],
           columns=('name', 'type', 'AvgBill')
                 )
df
```

#### **Method 1: Pandas Rename**

**pandas.DataFrame.rename(columns={"old_column_name" : "new_column_name"})**

- Change the column name **'AvgBill' to 'Bill'.**

- You need t0 tell pandas that you want to change your columns so you'll need to specify **axis=1 or columns=your_mapper**.

- I prefer setting **'columns='** because I **don't need to remember** if columns is **axis=1 or 0.**

1.JPG

```python
df.rename({'AvgBill' : 'Bill'}, axis=1)
```

```python
df.rename(columns={'AvgBill' : 'Bill'})
```

#### **Method 2: change column names via .rename() using function mapper**

```python
#  upper each of my column names

df.rename(columns=str.upper)
```

```python
# remove the first letter from each of my column names

df.rename(columns=lambda x: x[1:])
```

#### **Method 3: Pandas .columns attribute**

**pandas.DataFrame.columns = ['your', 'new', 'column', 'names']**

```python
df.columns = ['Name', 'type', 'ABill']
df
```

#### **Method 4 – Pandas .set_axis(axis=1)**

**pandas.DataFrame.set_axis(['your', 'new', 'column', 'names'2], axis=1)**

```python
df.set_axis(["Name", "Type", "AvgBill"], axis=1)
```
