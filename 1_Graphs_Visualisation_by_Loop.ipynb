```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("/content/drive/MyDrive/CloudyML/titanic.zip")
train.head()
```

```python
numeric_cols_train = train.select_dtypes(exclude=object)
display(numeric_cols_train.head())
```

```python
plt.figure (figsize = (20,15))
sns.set_style ('darkgrid')
for i, feature in enumerate (numeric_cols_train):
    plt.subplot (4,2, i*1 + 1 )
    sns.histplot (data = train, x = train[feature])
```

```python
plt.figure (figsize = (20,15))
sns.set_style ('darkgrid')
for i, feature in enumerate (numeric_cols_train):
    plt.subplot (4,2, i*1 + 1 )
    sns.boxplot (data = train, x = train[feature])
```
