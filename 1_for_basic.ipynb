### **List**

```python
x = ['navin', 65, 2.5]

for i in x:
  print(i)
```

```python
for i in [2,6,'Paul']:
  print(i)
```

```python
CityList = ['Hyderabad', 'Vizag', 'Bangalore', 'Chennai']

for city in CityList:
  print(city)
```

```python
for i in [10,20,30,40]:
  print(i)
```

### **Range**

```python
for i in range(10):
  print(i)
```

```python
for i in range(11,21,1):
  print(i)
```

### **Break**

```python
CityList = ['Hyderabad', 'Vizag', 'Bangalore', 'Chennai']

for city in CityList:
  print(city)
  if city == "Bangalore":
    break
```

### **Continue**

```python
CityList = ['Hyderabad', 'Vizag', 'Bangalore', 'Chennai']

for city in CityList:
  if city == "Bangalore":
    continue
  print(city)
```

### **Else**

```python
CityList = ['Hyderabad', 'Vizag', 'Bangalore', 'Chennai']

for city in CityList:
  print(city)
else:
  print("City list completed")
```

### **Nested Loop**

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```

### **if else**

```python
for i in [10,20,30,40]:
  if i>=25:
    print(i, "greater than 25")
  else:
    print(i, "less than 25")
```
