```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets/Telco-Customer-Churn.zip")
df.head()
```

### **Subplots**

```python
# Sub plot

plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
sns.countplot('Partner', hue = 'Churn', data = df)

plt.subplot(1,2,2)
sns.countplot('Dependents', hue = 'Churn', data = df)

plt.show()
```

### **Enumerate**

```python
df.columns
```

```python
feature = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
           'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
           'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
           'Contract', 'PaperlessBilling', 'PaymentMethod']
```

```python
list(enumerate(feature))
```

```python
plt.figure(figsize = (15, 30))
for i in enumerate(feature):
    plt.subplot(6, 3,i[0]+1)
    sns.countplot(i[1], hue = 'Churn', data = df)
    plt.xticks(rotation = 45)

plt.show()
```

```python
fig, ax = plt.subplots(len(feature), 1, figsize = (7, 50))   #a,b can be set by you

fig.tight_layout(pad=3.0)

for idx, col in enumerate(feature, 0):
    ax[idx].plot(df['tenure'], df[col], 'o')
    ax[idx].set_xlabel('tenure')
    ax[idx].set_title(col)

plt.show()
```
