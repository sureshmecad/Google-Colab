### **Types of Cross Validation**

#### 1) Non-Exhaustive Technique

- a) Hold-Out

- b) K-Fold

- c) Stratified K-Fold

#### 2) Exhaustive Techniques

- a) Leave One Out Cross Validation

- b) Leave P Out Cross Validation

##### **LOOCV : Leave One Out Cross Validation**
- model trains from 99% of data and predicts on 1% data

- **Disadvantage:** Since **one** record taking out for test so there is high possibility of **high variance**.

#### **KFold**
- **Adv:** Model is using all the data for training
- **Low variance** (in LOOCV, test size 1; in KFold, test size 10)
