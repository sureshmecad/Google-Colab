```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4]
y1 = [50, 20, 30, 10, 40]
y2 = [40, 50, 20, 30, 10]

plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
x = [0,1,2,3,4]
y1 = [50, 20, 30, 10, 40]
y2 = [40, 50, 20, 30, 10]
y3 = [30, 20, 50, 40, 10]
y4 = [40, 20, 10, 50, 30]

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

### **Adding titles**

```python
x = [0,1,2,3,4]
y1 = [50, 20, 30, 10, 40]
y2 = [40, 50, 20, 30, 10]
y3 = [30, 20, 50, 40, 10]
y4 = [40, 20, 10, 50, 30]

plt.figure(figsize=(7,6))
plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("PLOT-1")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("PLOT-2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("PLOT-3")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("PLOT-4")

plt.show()
```

### **Adding suptitle**

```python
x = [0,1,2,3,4]
y1 = [50, 20, 30, 10, 40]
y2 = [40, 50, 20, 30, 10]
y3 = [30, 20, 50, 40, 10]
y4 = [40, 20, 10, 50, 30]

plt.figure(figsize=(7,6))

plt.suptitle("MULTIPLE PLOTS")

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("PLOT-1")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("PLOT-2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("PLOT-3")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("PLOT-4")

plt.show()
```
