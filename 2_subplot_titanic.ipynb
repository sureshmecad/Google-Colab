```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/CloudyML/titanic.zip")
df.head()
```

```python
plt.style.use('ggplot')
fig, (ax1,ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize=(15,8))
plt.subplots_adjust(wspace=.5)

ax1.barh(df['Pclass'], df['SibSp'])
ax1.set_title("Pclass Vs Sibsp")

ax2.scatter(df['Age'], df['Fare'])
ax2.set_title("Age Vs Fare")

ax3 = sns.barplot(df['Age'], df['Fare'], data=df, hue='Sex')
ax3.set_title("Age Vs Fare")
```
