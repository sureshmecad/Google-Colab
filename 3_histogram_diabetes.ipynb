```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/CloudyML/diabetes.zip")
data.head()
```

### **Plotting Histogram  of all columns**

```python
data.hist(figsize=(10,10), bins=10);
```

### **Plotting Histogram of a particular column**

```python
data.hist(column="Glucose", figsize=(8,8), 
          xlabelsize=20, ylabelsize=20,
          xrot=45, bins=5, color='orange')
```

### **Plotting Histograms of multiple columns in a single plot**

```python
data1 = data[['BloodPressure','BMI']]
ax = data1.plot.hist(alpha=0.5, figsize=(10,10), bins=10, title = "BP VS BMI")
```

```python
data1.plot(kind='hist',
           alpha=0.7,
           bins=30,
           title='Histogram Of Test Scores',
           rot=45,
           grid=True,
           figsize=(12,8),
           fontsize=15, 
           color=['#A0E8AF', '#FFCF56'])

plt.xlabel('Test Score')
plt.ylabel("Number Of Students");
```
