```python
# Read Data
import numpy as np     # Linear Algebra (calculate the mean and standard deviation)
import pandas as pd    # manipulate data, data processing, load csv file I/O (e.g. pd.read_csv)

# Visualization
import seaborn as sns                  # Visualization using seaborn
import matplotlib.pyplot as plt        # Visualization using matplotlib
%matplotlib inline
```

```python
df = pd.read_csv("/content/drive/MyDrive/CloudyML/diabetes.zip")
```

```python
display(df.head(3))
display(df.tail(3))
```

```python
# Scatter plot between "Age" & "Glucose"
plt.figure(figsize=(15,10))
plt.scatter(df['Age'], df['Glucose'])

plt.xlabel('Age', fontsize=16, fontweight='bold')
plt.ylabel('Glucose', fontsize=16, fontweight='bold')

plt.title('Age Vs Glucose', fontsize=20, fontweight='bold')

plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)

plt.show()
```

```python
# Scatter plot between "cement" and "water"
fig = plt.figure(figsize=(13,8))
ax = fig.add_subplot(111)
plt.scatter(df["Age"], df["Glucose"], c=df["BloodPressure"], s=df["BMI"]*4,
            linewidth=1, edgecolor="k", cmap="viridis_r")
ax.set_facecolor("w")
ax.set_xlabel("Age")
ax.set_ylabel("Glucose")
lab = plt.colorbar()
lab.set_label("Healthcare")
plt.title("scatter plot between Age and Glucose")
plt.grid(True,alpha=.3)
plt.show()
```
