```python
# Method 1

list_of_num = []
for i in range(16):
  list_of_num.append(i)

print(list_of_num)
```

<h1 style="background-color:LimeGreen; font-family:newtimeroman; font-size:200%; text-align:center; border-radius: 15px 50px;"> 10) Get the difference between two Lists </h1>

```python
list1 = ['tanmay', 'jain', 'python', 'advance', 'projects']
list2 = ['tanmay', 'jain', 'python']

set1 = set(list1)
set2 = set(list2)

list3 = list(set1.symmetric_difference(set2))
print(list3)
```
